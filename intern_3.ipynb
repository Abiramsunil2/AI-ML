```python
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()
```

```python
label_encoder=LabelEncoder()
encoded_data['Gender_Label_Encoded']=label_encoder.fit_transform(encoded_data['Gender'])
```

```python
one_hot_encoded = pd.get_dummies(encoded_data['Department_mode_imputed'], prefix='Department_mode_imputed')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)
```

```python
Rating_mapping = {1: '1STAR', 2: '2STAR', 3: '3STAR', 4: '4STAR', 5: '5STAR'}
encoded_data['Rating_Mapped'] = encoded_data['Performance_Rating'].map(Rating_mapping)
```

```python
print(encoded_data[['Gender', 'Gender_Label_Encoded', 'Department', 'Department_mode_imputed_Sales', 'Department_mode_imputed_IT', 'Department_mode_imputed_HR', 'Department_mode_imputed_Marketing', 'Performance_Rating', 'Rating_Mapped']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
numerical_col=['Age','Experience_Years','Monthly_Salary']
df.dropna(subset=numerical_col,inplace=True)
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
df['Age_MinMax']=minmax_scaler.fit_transform(df[['Age']])
df['Age_Standard']=standard_scaler.fit_transform(df[['Age']])
```

```python
df['Experience_Years_MinMax']=minmax_scaler.fit_transform(df[['Experience_Years']])
df['Experience_Years_Standard']=standard_scaler.fit_transform(df[['Experience_Years']])
```

```python
df['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(df[['Monthly_Salary']])
df['Monthly_Salary_Standard']=standard_scaler.fit_transform(df[['Monthly_Salary']])
```
